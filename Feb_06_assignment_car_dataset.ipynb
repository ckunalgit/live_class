### Dump all data into car table

```python
sql = "INSERT INTO cardataset.car(buying,maint,doors,persons,lug_boot,safety) VALUES (%s, %s, %s, %s, %s, %s)"
```

```python
import csv 
with open('car.data','r') as f:
    car_data = csv.reader(f,delimiter = '\n')
    for i in car_data:
        val = tuple(i[0].split(','))
        print(sql,val)
        cursor.execute(sql,val)
```
